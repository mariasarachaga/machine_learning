```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = "/kaggle/input/tp3-ml/"
y_images = np.load(f'{DATA_PATH}y_images.npy')

unique_classes = np.unique(y_images)
counts = np.bincount(y_images.astype(int))

plt.figure(figsize=(12,6))
plt.bar(unique_classes, counts[:len(unique_classes)], color='skyblue', edgecolor='black')
plt.xlabel("Clase")
plt.ylabel("Cantidad de muestras")
plt.xticks(unique_classes)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
